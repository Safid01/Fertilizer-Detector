# Synthetic Fertilizer Dataset Generator
This notebook generates a synthetic fertilizer recommendation dataset for rice crops in Bangladesh.

```python
import numpy as np
import pandas as pd
import random
```

```python
# Crop types in Bangladesh
crops = ["Boro rice", "Aman rice", "Aus rice"]
```

```python
# Function to generate fertilizer recommendation
def fertilizer_recommendation(shi, crop):
    # Base ranges from BRRI guidelines
    urea = np.random.randint(70, 141)  # 70–140
    tsp = np.random.randint(30, 81)   # 30–80
    mop = np.random.randint(20, 61)   # 20–60

    # Adjust by SHI
    if shi < 50:
        urea *= 1.2
        tsp *= 1.2
        mop *= 1.2
    elif shi > 75:
        urea *= 0.85
        tsp *= 0.85
        mop *= 0.85

    return round(urea, 1), round(tsp, 1), round(mop, 1)
```

```python
# Generate synthetic dataset
def generate_dataset(n_samples=500):
    data = []
    for _ in range(n_samples):
        shi = np.random.randint(30, 91)  # SHI 30–90
        N = round(np.random.uniform(5, 20), 2)   # mg/kg
        P = round(np.random.uniform(3, 15), 2)
        K = round(np.random.uniform(20, 60), 2)
        pH = round(np.random.uniform(4.5, 8.0), 2)
        crop = random.choice(crops)

        urea, tsp, mop = fertilizer_recommendation(shi, crop)

        data.append([shi, N, P, K, pH, crop, urea, tsp, mop])

    df = pd.DataFrame(data, columns=["SHI", "N", "P", "K", "pH", "Crop",
                                     "Urea (kg/ha)", "TSP (kg/ha)", "MoP (kg/ha)"])
    return df
```

```python
# Generate and save dataset
df = generate_dataset(500)
df.to_csv("synthetic_rice_fertilizer_dataset.csv", index=False)

print("✅ Dataset generated: synthetic_rice_fertilizer_dataset.csv")
```
